# Importing the required basics
In this example we'll make our first plot using Matplotlib; Python's most commonly-used data visualisation library.
Our first graph will contain a single period of a sine and a cosine, so we'll need `math`, and of course `matplotlib` itself! For most graphs we'll use a submodule from Matplotlib called `pyplot`, which provides a MATLAB-like plotting interface.

```python
from math import sin, cos, pi
import matplotlib.pyplot as plt
```

# Welcome and initialisation
After our required imports, of course we start with a nice welcome message, followed by the initialisation of our variables.
In an XY-plot, each data point in a graph consists of an x and a y value. Pyplot expects the individual coordinates in separate sequences, so we'll create lists for x, sin(x), and cos(x), respectively:

```python
print('Welcome to my first plotting program')
print('This program will plot one period of the sine and cosine function in one graph')

# Definition of our step size delta x, and the number of datapoints:
n = 100
step = 2.0 * pi / n

# Initialisation of x and y value lists
xtab = []
y1tab = []
y2tab = []
```

# Populating our lists with x, y data
The next step is to fill our lists with ranges of x, cos(x) and sin(x) data. Since we want to perform the same calculation for a fixed number of times we'll use a `for`-loop for this. Because the `range()` function only works with integers we need to do some scaling to generate our x-values:

```python
# Note how we use an integer range function to generate floats
# Since the range function omits the final value given as end, we add 1
# to make sure our last value for x is exactly 2*pi
for i in range(n + 1):
    x = i * step
    y1 = sin(x)
    y2 = cos(x)

    xtab.append(x)
    y1tab.append(y1)
    y2tab.append(y2)
```

# Plotting the data
With our data created in three lists the only step left is to actually plot the data! We can do this with the `plot()` function from `pyplot`. To actually reveal the generated plot window we should call `show()`.

```python
# Plot the sine and cosine lines in a graph
plt.plot(xtab, y1tab)
plt.plot(xtab, y2tab)

# Add a title, axis labels, and a legend
plt.title('One period of the sine and cosine functions')
plt.xlabel('x [rad]')
plt.ylabel('y')
plt.legend(('sin(x)', 'cos(x)'))

# Don't forget to call show()!
plt.show()
```

More options, such as the ability to put multiple sub-plots in one figure, are available if we explicitly define a figure (with `plt.figure()`) before our other plot commands. We can for instance do the following:

```python
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(xtab, y1tab)
plt.xlabel('x [rad]')
plt.ylabel('sin(x)')
plt.subplot(2, 1, 2)
plt.plot(xtab, y2tab)
plt.xlabel('x [rad]')
plt.ylabel('cos(x)')
plt.show()
```
